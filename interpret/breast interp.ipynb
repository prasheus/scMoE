```python
import os
```

```python
adatas = []
for name, paths in study_files.items():
    matrix_path = paths["matrix"]
    genes_path = paths["genes"]
    cells_path = paths["cells"]
    if not os.path.exists(matrix_path) or not os.path.exists(genes_path) or not os.path.exists(cells_path):
        print(f"{name}: missing file")
        continue

    ad = sc.read_mtx(matrix_path).T  

    gene_df = pd.read_csv(genes_path, header=None, sep="\t")
    ad.var_names = gene_df.iloc[:, 0].tolist()

    cells_df = pd.read_csv(cells_path, index_col=0)
    assert len(cells_df) == ad.n_obs, f"barcode mismatch in {name}: {len(cells_df)} barcodes vs {ad.n_obs} cells"
    ad.obs_names = cells_df.index
    ad.obs = ad.obs.join(cells_df, how="left")

    ad.obs["study"] = name
    ad.obs["batch"] = name
    ad.obs["matrix_units"] = paths["units"]

    adatas.append(ad)
```

```python
# join failed (checked for duplicates)
for i, ad in enumerate(adatas):
    dup_genes = ad.var_names.duplicated().sum()
    if dup_genes > 0:
        print(f"{ad.obs['study'].iloc[0]}: {dup_genes} duplicate gene name")
```

```python
# not overwriting gene names as every gene names has expression information
for ad in adatas:
    ad.var_names = ad.var_names.astype(str) #string
    # fucntion adds like _1, _2, etc. to duplicates
    ad.var_names_make_unique()
```

```python
common_genes = adatas[0].var_names
for i, ad in enumerate(adatas[1:], 1):
    common_genes = common_genes.intersection(ad.var_names)
```

```python
# subset the data only to common genes
```

```python
adatas_filtered = []
for ad in adatas:
    ad_sub = ad[:, common_genes].copy()
    adatas_filtered.append(ad_sub)
    study_name = ad_sub.obs['study'].iloc[0]
```

```python
del adatas; import gc; gc.collect()
```

```python
adata_all = adatas_filtered[0].concatenate(
    adatas_filtered[1:],
    batch_key="batch",
    join="inner",
    index_unique="-"
)
```

```python
adata_all.n_vars #genes
```

```python
# checkpoint
for col in adata_all.obs.columns:
    if adata_all.obs[col].dtype == 'object':
        adata_all.obs[col] = adata_all.obs[col].astype(str)
adata_all.write_h5ad('breast_cancer_3CA_combined.h5ad')
```

```python
#quality control

# https://www.weizmann.ac.il/sites/3CA/methods
```

```python
# gene distribution
print((adata_all.obs['n_genes_by_counts'] < 200).sum()) #cells having genes less than 200
print((adata_all.obs['n_genes_by_counts'] > 6000).sum()) #cells having genes less than 6000
```

```python
# filtering genes
min_genes = 200
max_genes = 6000
keep = (adata_all.obs['n_genes_by_counts'] >= min_genes) & (adata_all.obs['n_genes_by_counts'] <= max_genes)
adata_qc = adata_all[keep, :].copy()
```

```python
del adata_all; gc.collect()
```

```python
# NORMALIZATION
sc.pp.normalize_total(adata_qc, target_sum=1e4)
sc.pp.log1p(adata_qc)
```

```python
# checkpoint 
for col in adata_qc.obs.columns:
    if adata_qc.obs[col].dtype == 'object':
        adata_qc.obs[col] = adata_qc.obs[col].astype(str)

adata_qc.write_h5ad('breast_cancer_3CA_clean.h5ad')
```

```python
# CHECKING FOR BATCH EFFECT
# get highly variable genes

sc.pp.highly_variable_genes(adata_qc, n_top_genes=2000, batch_key='batch')
```

```python
adata_umap = adata_qc[:, adata_qc.var.highly_variable].copy() #copy for batch test
```

```python
# dim reduction and UMAP
sc.tl.pca(adata_umap, svd_solver='arpack')
# knn
sc.pp.neighbors(adata_umap, n_neighbors=15, n_pcs=30)
sc.tl.umap(adata_umap)
```

```python
sc.pl.umap(adata_umap, color=['study', 'cell_type'], palette='tab20', wspace=0.4)
```
