```python
import scanpy as sc 
import pandas as pd
```

```python
adata = sc.read_mtx('Data_Wu2021_Breast/Exp_data_UMIcounts.mtx')
```

```python
genes = pd.read_csv('Data_Wu2021_Breast/Genes.txt', header=None, sep='\t').iloc[:,0].tolist()
```

```python
adata = adata.T
adata.var_names = genes
```

```python
# combining with cell metadata (barcode as index)
cells = pd.read_csv('Data_Wu2021_Breast/Cells.csv', index_col=0)
```

```python
assert len(cells) == adata.n_obs

# barcodes should match cells
```

```python
adata.obs_names = cells.index
```

```python
adata.obs = adata.obs.join(cells, how='left')
```

```python
adata.obs['study'] = 'Wu2021'
```
